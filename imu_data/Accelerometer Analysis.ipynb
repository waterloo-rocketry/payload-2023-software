```python
#Time Intervals are taken from visual inspection of graphs to approximate time interval of actual turn
#There are a total of 4 pitch turns (90 degrees each) in this graph, 9500-10500 should be a fast 90 degree turn
accel_data2 = accel_data[((accel_data['timestamp'] > 2000) & (accel_data['timestamp'] < 5000))]
accel_data2["Accel Sum"] = accel_data2["Accel X"] + accel_data2["Accel Y"] + accel_data2["Accel Z"]
print(accel_data2)
```

```python
#standardization of units to dps (Run me once only!)

#timestamp is in ms, so convert to s
#Accel data values have typical value of 0.244mg/LSB, since CTRL1_XL is set to 0x3c, i.e. 8g, 52HZ setting
#So the conversion rate is 0.244 (Typical Value) * 0.001 (mg to g) * 9.81 (g to m/s^2) = 

#371 is referring to row 0 actually
accel_data2 = accel_data[((accel_data['timestamp'] > 2000) & (accel_data['timestamp'] < 5000))]
print(accel_data2)
accel_data2["timestamp"] = (accel_data2["timestamp"] - accel_data2.loc[371]["timestamp"])/ 1000
accel_data2["Accel X"] = accel_data2["Accel X"] * 0.00239
accel_data2["Accel Y"] = accel_data2["Accel Y"] * 0.00239
accel_data2["Accel Z"] = accel_data2["Accel Z"] * 0.00239
accel_data2["Accel Sum"] = accel_data2["Accel X"] + accel_data2["Accel Y"] + accel_data2["Accel Z"]
print(accel_data2)
```

```python
accel_data2.plot(x="timestamp", y="Accel Z")
```
