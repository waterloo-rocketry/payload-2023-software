```python
import pandas as pd
```

```python
data = pd.read_csv("./gyro_pitch.csv")

print (data)
```

```python
gyro_data = data[["timestamp","Gyro X", "Gyro Y", "Gyro Z"]]
print (gyro_data)
```

```python
#Run me only once please (Please go gentle on me uwu)
#Seriously though, since this is directly modifying data on dataset, please do only run once
#A red warning window will pop up when ran, but it can be ignored
gyro_data["Gyro X"] = gyro_data["Gyro X"] - gyro_data.loc[0]["Gyro X"]
gyro_data["Gyro Y"] = gyro_data["Gyro Y"] - gyro_data.loc[0]["Gyro Y"]
gyro_data["Gyro Z"] = gyro_data["Gyro Z"] - gyro_data.loc[0]["Gyro Z"]
gyro_data["timestamp"] = gyro_data["timestamp"] - gyro_data.loc[0]["timestamp"]
gyro_data.plot(x="timestamp", y="Gyro X")
```

```python
gyro_data["Gyro Sum"] = gyro_data["Gyro X"] + gyro_data["Gyro Y"] + gyro_data["Gyro Z"]
```

```python
#Time Intervals are taken from visual inspection of graphs to approximate time interval of actual turn
#There are a total of 4 pitch turns (90 degrees each) in this graph, 9500-10500 should be a fast 90 degree turn
gyro_turn2 = gyro_data[((gyro_data['timestamp'] > 9500) & (gyro_data['timestamp'] < 10500))]
print(gyro_turn2)
gyro_turn2.plot(x="timestamp", y="Gyro Sum")
```

```python
#standardization of units to dps (Run me once only!)

#timestamp is in ms, so convert to s
#gyro data values have conversion rate of 0.07, since CTRL2_G is set to 0x4c, i.e. 1000 dps setting
#So the conversion rate is 70 (Typical Value) * 0.001 (mdps to dps)

#1750 is referring to row 0 actually
gyro_turn2["timestamp"] = (gyro_turn2["timestamp"] - gyro_turn2.loc[1750]["timestamp"])/ 1000
gyro_turn2["Gyro X"] = gyro_turn2["Gyro X"] * 0.07
gyro_turn2["Gyro Y"] = gyro_turn2["Gyro Y"] * 0.07
gyro_turn2["Gyro Z"] = gyro_turn2["Gyro Z"] * 0.07
gyro_turn2["Gyro Sum"] = gyro_turn2["Gyro Sum"] * 0.07
print(gyro_turn2)
```

```python
r_sum = 0 #running riemann sum
t_sub_1 = 0 #t value at the last data point

for index, row in gyro_turn2.iterrows():
    r_sum += row["Gyro Sum"] * (row["timestamp"] - t_sub_1)
    #rationale : riemann sum between each two data points in the sample data
    
    t_sub_1 = row["timestamp"]
print(r_sum) #this should be the riemann sum of degrees turned in the 1 second time
```
